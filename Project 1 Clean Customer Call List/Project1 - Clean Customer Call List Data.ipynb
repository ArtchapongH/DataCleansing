```python
# import library

import pandas as pd
```

```python
# get dataframe

df = pd.read_excel(r'D:\Python Project\Project1\Customer Call List.xlsx')

df
```

```python
# drop duplicate rows

df=df.drop_duplicates()

df
```

```python
# drop unuseful column

df.drop(columns='Not_Useful_Column', inplace=True)

df.head()
```

```python
# see what are characters needed to be got rid from 'Last_Name'

df['Last_Name'].str.replace('[a-zA-Z]','').unique()
```

```python
# filter only characters need to be got rid

tmp = df['Last_Name'].str.replace('[a-zA-Z]','').unique()

charList = []
for c in tmp:
    if not (c in ['','  ']) and not(pd.isnull(c)):
        charList.append(c)
        
        
charList
```

```python
# because some of them are not only 1 character so, we need to do some modification to get only 1 character of each

char_to_del=[]
for c in charList:
    if len(c)==1:
        char_to_del.append(c)
    else:
        char_to_del = char_to_del + list(c)
        

char_to_del = list(set(char_to_del))

char_to_del
```

```python
# join each characters to be one text to pass to strip method at once

char_to_del = ''.join(char_to_del)

char_to_del
```

```python
# strip all unwanted characters to strip method to clean 'Last_Name' column

df['Last_Name'] = df['Last_Name'].str.strip(char_to_del)

df
```

```python
# get rid of non numuric characters from 'Phone_Number' column

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')

df
```

```python
# format phone number

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x)).apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])

df
```

```python
# clean non phone number by replace with blank

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: '' if x[-2:]=='--' else x)

df
```

```python
# split 'Address' column to 'Street_Address' , 'State' , and 'Zip_Code'  columns

df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',2,expand=True)


df
```

```python
# replace 'Yes' and 'No' to 'Y' and 'N' in column 'Paying Customer'

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','Y')

df
```

```python
# replace 'Yes' and 'No' to 'Y' and 'N' in column 'Do_Not_Contact'

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','Y')

df
```

```python
# replace all 'N/A' and null value in dataframe with blank
 
df = df.replace('N/a','')


df = df.fillna('')


df
```

```python
# drop rows that 'Do_Not_Contact' columns contains value 'Y'

df = df[df['Do_Not_Contact']!='Y']
        
df
```

```python
# drop rows that 'Do_Not_Contact' columns contains value 'Y'

df = df[df['Phone_Number']!='']
        
df
```

```python
# reset index column

df = df.reset_index(drop=True)

df
```
