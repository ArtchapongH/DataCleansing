```python
# import library

import pandas as pd
```

```python
# read csv file into a pandas df with the specified column name

columns_name=['SalesOrderNumber','SalesOrderLine','OrderDate','CustomerName','EmailAddress','Item','Quantity','UnitPrice','TaxAmount']
df = pd.read_csv(r'D:\Python Project\Project2\Sales Order 2021.csv',header=None,names=columns_name)

df.head(10)
```

```python
# get dataframe information

df.info()
```

```python
# check for duplicates

duplicate_rows = df[df.duplicated()]

if not duplicate_rows.empty:
    print('Dupplicate rows: ',duplicate_rows)
else:
    print('No duplicate rows')
```

```python
# split the 'OrderDate' column into seperate 'Year' and 'Month' columns

df[['Year','Month','Day']] = df['OrderDate'].str.split('-', expand=True)

df.head()
```

```python
# insert the new 'Year' and 'Month' columns after the 'OrderDate' column

# get the index of the 'OrderDate' column

order_date_index = df.columns.get_loc('OrderDate')


# insert 'Year' and 'Month' columns after 'OrderDate'

df.insert(order_date_index+1, 'Year', df.pop('Year'))
df.insert(order_date_index+2, 'Month', df.pop('Month'))

df.head()
```

```python
# drop 'Day' columns

df.drop('Day', axis=1, inplace=True)

df.head()
```

```python
# split the 'CustomerName' column into seperate 'FirstName' and 'LastName' columns

df[['FirstName','LastName']] = df['CustomerName'].str.split(' ', n=1, expand=True)

df.head()
```

```python
# insert the LastName and FirstName columns after the 'CustomerName' column

# get the index of the 'OrderDate' column

customername_index = df.columns.get_loc('CustomerName')


# insert 'Year' and 'Month' columns after 'OrderDate'

df.insert(customername_index+0, 'FirstName', df.pop('FirstName'))
df.insert(customername_index+1, 'LastName', df.pop('LastName'))

df.head()
```

```python
# drop 'CustomerName' columns

df.drop('CustomerName', axis=1, inplace=True)

df.head()
```

```python
# save transformed dataframe as csv file

df.to_csv(r'D:\Python Project\Project2\transformed_data.csv', index=False)
```
